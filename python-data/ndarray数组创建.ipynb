##ndarray数组创建

```python
import numpy as np
from skimage.util import dtype
```

```python
#基础创建方法
arr = np.array([1,2,3])
print(arr.ndim)  #属性
print(arr)
```

```python
#copy
arr1 = np.copy(arr)
print(arr1)
arr1[0] = 8
print(arr1)
print(arr)
```

```python
#预定义形状
#全0  全1  未初始化
#全0
arr = np.zeros((2,3),dtype=int)
print(arr)
print(arr.dtype)
```

```python
arr = np.zeros((2,),dtype=int)
print(arr)
```

```python
#全1
arr = np.ones((5,8),dtype=int)
print(arr)
```

```python
#未初始化
arr = np.empty((5,8))
print(arr)
```

```python
arr = np.full((3,4),2026)
print(arr)
```

```python
arr1 = np.zeros_like(arr)
print(arr1)
```

```python
arr1 = np.ones_like(arr)
print(arr1)
```

```python
arr1 = np.empty_like(arr)
print(arr1)
```

```python
#等差数列
arr = np.arange(1,10,1)
print(arr)
```

```python
#等间隔数列
arr = np.linspace(0,100,5)
print(arr)
```

```python
arr = np.linspace(0,101,5,dtype=int)
print(arr)
arr = np.arange(0,101,25)
print(arr)
```

```python
#对数间隔数列
arr = np.logspace(0,4,2,base=2)
print(arr)
```

```python
#特殊矩阵
#单位矩阵：主对角线上的数字为1，其他的数字为0
arr = np.eye(3,4,dtype=int)
print(arr)
```

```python
#对角矩阵：主对角线上非零的0，其他的数字为0
arr = np.diag([1,2,3])
print(arr)
```

```python
#随机数组的生成
#生成0-1之间的随机浮点数（均匀分布）
arr = np.random.randn(3,4)
print(arr)
```

```python
#生成指定范围区间的随机浮点数
arr = np.random.uniform(3,8,(3,4))
print(arr)
```

```python
#生成指定范围区间的随机整数
arr = np.random.randint(3,30,(3,4))
print(arr)
```

```python
#生成随机数列（正态分布）
#两边小，中间大
arr = np.random.randn(3,4)
print(arr)
```

```python
#设置随机种子
np.random.seed(200)
arr = np.random.randint(1,10,(2,5))
print(arr)
```
