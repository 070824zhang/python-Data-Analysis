```python
#缺失值的处理
import pandas as pd
import numpy as np
s = pd.Series([1,2,np.nan,None,pd.NA])
df = pd.DataFrame([[1,pd.NA,2],[2,3,5],[None,4,6]],columns=['第一列','第二列','第三列'])
print(s)
#检查看是否是缺失值
print(s.isna())
print(s.isnull())
print(df.isna())
print(df.isnull())
#缺失值的个数
print(df.isna().sum())
print(s.isnull().sum())
```

```python
#剔除缺失值
print(s.dropna())
print('-'*30)
print(df)
print(df.dropna())#剔除一整条的记录
print(df.dropna(how='all'))#如果所有的值都是缺失值，删除这一行
print(df.dropna(thresh=2))#至少有n个值不是缺失值，然后就保留
print(df.dropna(axis=1))#剔除一整列的记录
print(df.dropna(subset=['第一列']))#如果某列有缺失值，则删除这一行
```

```python
#填充缺失值
df = pd.read_excel('students.xlsx')
print(df)
print(df.isna().sum(axis=0))
print(df.fillna({'id':0}))#填充指定值
print(df.fillna(df[['id']].mean()))#填充平均值
print(df.ffill())#用前面的相邻值填充
print(df.bfill())#用后面的相邻值进行填充
```
