```python
#数据分箱  pd.cut(x,bins,labels)
import pandas as pd
df = pd.read_excel('students.xlsx')
df.head()
```

```python
df1 = df.head(2)[['id','姓名']]
df1
```

```python
pd.cut(df['id'],bins=2).value_counts()#bins=整数，分成n段区间,起始值，结束值是所有数据的最小值·最大值
df['年龄阶段'] = pd.cut(df['年龄'],bins=[19,19.5,20,21,22.5],labels=['年轻','中年','壮年','少年'])#bins=list,分成n段区间
pd.qcut(df['年龄'],1).value_counts()#等频率分
```
