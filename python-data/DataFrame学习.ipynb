##dataframe的创建方式

```python
import pandas as pd
import numpy as np
#通过series来创建
s1 = pd.Series([1,2,3,4,5,6])
s2 = pd.Series([4,5,6,7,8,9])
df = pd.DataFrame({"第一列：":s1,"第二列":s2})
df
```

```python
#通过字典来创建
df = pd.DataFrame(
    {
        "name":["tom","jack","alice","bob","allen"],
        "age":[10,20,23,29,50],
        "score":[60.5,80,30.6,70,83.3]
    },index=[1,2,3,4,5],columns=["name","score","age"]
)
df
```

```python
#dataframe的属性
print('行索引：')
print(df.index)
print('列索引：')
print(df.columns)
print('值：')
print(df.values)
```

```python
print('维度：',df.ndim)
print('形状：',df.shape)
print('元素个数：',df.size)
print('数据类型:')
print(df.dtypes)
```

```python
#行列转置
print(df.T)
```

```python
#获取元素  loc iloc at iat
#获取某行
print(df.loc[4])
print(df.iloc[3])
```

```python
#获取某列
print(df.loc[:,'name'])
print(df.iloc[:,0])
```

```python
#单个元素
print(df.at[3,'score'])
print(df.iloc[2,1])
print(df.loc[3,'score'])
print(df.iloc[2,1])
```

```python
#获取单列数据
print(df['name'])
print(df.name)
print(df[['name','age']])
```

```python
#查看部分数据
print(df.head(2))
```

```python
#使用布尔索引筛选数据
df[df.score>70]
df[ (df.score>70)&(df.age<20)]
```

```python
#随机抽样
df.sample()
```
