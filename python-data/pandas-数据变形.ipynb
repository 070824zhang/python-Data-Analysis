```python
#数据变形
import pandas as pd
data = {
    'id':[1,2],
    'name':['alice','bob'],
    'Math':[90,85],
    'English':[88,92],
    'History':[95,89],
}
df = pd.DataFrame(data)
df.T
#宽表转换成长表
df2 = pd.melt(df,id_vars=['id','name'],var_name='科目',value_name='分数')
df2.sort_values('name')
#长表转宽表
pd.pivot(df2,index=['id','name'],columns='科目',values='分数')
```
