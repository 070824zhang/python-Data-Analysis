```python
import pandas as pd
import numpy as np
df = pd.DataFrame(
    {
        "name":["tom","jack","alice","bob","allen"],
        "age":[10,20,23,29,50],
        "score":[60.5,80,30.6,70,83.3]
    },index=[1,2,3,4,5],columns=["name","score","age"]
)
df
```

```python
print(df.head())
print(df.tail(1))
```

```python
print(df.isin(['jack']))#查看元素是否包含在元素集合中
```

```python
print(df.isna())
```

```python
print(df['score'].sum())
print(df.score.max())
print(df.score.min())
print(df.score.mean())
print(df.score.median())
print(df.score.mode())
```

```python
print(df.score.std())#标准差
print(df.score.var())#方差
print(df.score.quantile(0.25))#分位数
```

```python
print(df.describe())
```

```python
print(df.duplicated(subset=['age']))#检查重复
```
